## Product Distribution

This analysis identifies the frequency of each product in the dataset.

```python
df["Product"].value_counts()
```

```python
df["Product"].value_counts().plot(kind="bar", figsize=(10,5))
plt.title("Product Distribution")
plt.xlabel("Products")
plt.ylabel("Frequency")
plt.show()
```

## Payment Method Analysis

This analysis helps identify the most commonly used payment methods by customers.

```python
df["PaymentMethod"].value_counts()
```

```python
df["PaymentMethod"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%"
)
plt.title("Payment Method Distribution")
plt.ylabel("")
plt.show()
```

## Order Status Analysis

This section analyzes the distribution of order statuses.

```python
df["OrderStatus"].value_counts()
```

```python
df["OrderStatus"].value_counts().plot(kind="bar")
plt.title("Order Status Distribution")
plt.xlabel("Status")
plt.ylabel("Count")
plt.show()
```

## Referral Source Analysis

Referral sources indicate which channels are driving customer traffic.

```python
df["ReferralSource"].value_counts()
```

```python
df["ReferralSource"].value_counts().plot(kind="bar")
plt.title("Referral Source Distribution")
plt.xlabel("Source")
plt.ylabel("Count")
plt.show()
```

## Sales Trend Analysis

This section examines sales performance over time to identify trends.

```python
df["Date"] = pd.to_datetime(df["Date"])
df["Month"] = df["Date"].dt.month
```

```python
monthly_sales = df.groupby("Month")["TotalPrice"].sum()
monthly_sales
```

```python
monthly_sales.plot(marker="o")
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.grid(True)
plt.show()
```

## Outlier Detection

Boxplots help identify unusual values that significantly differ from the rest of the data.

```python
plt.boxplot(df["TotalPrice"])
plt.title("Outlier Detection for TotalPrice")
plt.show()
```

## Correlation Analysis

Correlation analysis helps determine relationships among numerical variables.

```python
corr_matrix = df.select_dtypes(include=["int64","float64"]).corr()
corr_matrix
```

```python
plt.figure(figsize=(8,6))
plt.imshow(corr_matrix)
plt.colorbar()
plt.xticks(range(len(corr_matrix.columns)), corr_matrix.columns, rotation=90)
plt.yticks(range(len(corr_matrix.columns)), corr_matrix.columns)
plt.title("Correlation Matrix")
plt.show()
```
